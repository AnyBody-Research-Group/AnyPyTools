# Using AnyBody from Python 

***

The package **`anypytools`** has many utilities and tools to work with the AnyBody Modeling System (AMS). 

This tutorial includes the following:

### Topics

* [Getting started](01 - Getting started with anypytools.ipynb)  Install and run simple AnyBody macros 
* [Generating macros](02 - Generating macros.ipynb) Generate and run complex macros
* [Working with AnyBody output](03 - Working with output from Anybody.ipynb) Loading AnyOutput files and HDF5 files generated by AMS
* [Batch processing](04 - Batch processing many model.ipynb) Techniques to batch process many models 
* [Advanced studies](05 - Monte Carlo studies etc.ipynb) Build macros for parameter, monte carlo and Latin hypercube studies.
